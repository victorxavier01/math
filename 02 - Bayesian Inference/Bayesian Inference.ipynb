# Bayesian Inference
Bayesian Inference is one of the most powerful frameworks in statistics and data science. It provides a mathematically rigorous way to update beliefs when new evidence appears.

## 1. The Core Idea
At its heart, Bayesian inference answers one question:
> How should we update what we believe after observing new data?

Instead of treating parameters as fixed unknown values (frequentist view), Bayesian statistics treats them as random variables with probability distributions.

This allows us to:

* Quantify uncertainty

* Incorporate prior knowledge

* Update beliefs sequentially

## 2. Bayes' Theorem

Bayes‚Äô Theorem is derived from **conditional probability**:
$$
P(A \mid B) = \frac{P(A \cap B)}{P(B)}
$$

$$
{P(A \cap B)} = P(A) + P(B) - P(A \cup B)
$$

### Bayesian Notation (More General Form)

In inference problems:
$$
P(\theta \mid D) = \frac{P(D \mid \theta) P(\theta)}{P(D)}
$$

Where:
* Œ∏ = parameter (what we want to estimate)
* D = observed data
---
#### The 4 Key Components
1. Prior ‚Äî $ ùëÉ(ùúÉ) $<br>Your belief about the parameter before seeing data.

2. Likelihood ‚Äî $ P(D‚à£Œ∏) $<br>How likely the data is given a parameter value.

3. Evidence (Marginal Likelihood) ‚Äî $ P(D) $<br>A normalizing constant:
$$
P(D) = \int P(D \mid \theta) P(\theta) d\theta
$$

4. Posterior ‚Äî $P(Œ∏‚à£D)$<br>Your updated belief after seeing data.

### 3. Intuition Example ‚Äî Biased Coin

Suppose we want to estimate the probability ùúÉ of getting heads.

We toss a coin 10 times:

* 8 heads
* 2 tails

#### Step 1: Define the Likelihood

The coin toss follows a Binomial distribution:
$$
P(D \mid \theta) = \binom{n}{k} \theta^k (1 - \theta)^{n-k}
$$

#### Step 2: Choose a Prior

A common choice is the Beta distribution:
$$
\theta \sim Beta(\alpha, \beta)
$$
The Beta distribution is:
$$
P(\theta) = \frac{\Gamma(\alpha + \beta)}
{\Gamma(\alpha)\Gamma(\beta)}
\theta^{\alpha - 1}(1 - \theta)^{\beta - 1}
$$
It is conjugate to the Binomial distribution.

#### Step 3: Compute the Posterior

If:
$$
\theta \sim Beta(\alpha, \beta)
$$
After observing k heads in n trials:
$$
\theta\mid D \sim Beta(\alpha + k, \beta + n - k)
$$

---
##### Example Calculation
Assume prior:
$$
Beta(2,2)
$$

This represents a mild belief that the coin is fair.

Data:

* 8 heads

* 2 tails

Posterior:
$$
Beta(2+8,2+2) = Beta(10,4)
$$
Posterior mean:
$$
E[\theta] = \frac{\alpha}{\alpha + \beta}
$$
$$
E[\theta] = \frac{10}{14}\approx0,714
$$
So instead of simply saying:
$$
\tilde{\theta} = 0.8
$$
We now say:
> We believe the true probability is around 0.714, with quantified uncertainty.

## 4. Python Implementation
Let‚Äôs visualize prior and posterior.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Prior parameters
alpha_prior = 2
beta_prior = 2

# Data
n = 10
k = 8

# Posterior parameters
alpha_post = alpha_prior + k
beta_post = beta_prior + n - k

x = np.linspace(0, 1, 1000)

plt.plot(x, beta.pdf(x, alpha_prior, beta_prior), label='Prior Beta(2,2)')
plt.plot(x, beta.pdf(x, alpha_post, beta_post), label='Posterior Beta(10,4)')
plt.legend()
plt.title("Bayesian Updating")
plt.show()
```

This graph clearly shows how the belief shifts after observing data.

## 5. Sequential Updating
Bayesian inference naturally supports streaming data.

After 10 tosses ‚Üí update
After 20 tosses ‚Üí update again

Posterior becomes the new prior.

This is extremely useful in:

* Online learning

* A/B testing

* Real-time risk estimation
